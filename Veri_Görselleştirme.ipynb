## Veri Görselleştirmesi

```python
# Tarih sütunlarını çıkarma
df_cleaned = df.drop(columns=['joining_date'])  # 'joining_date' sütunu gibi tarih formatındaki sütunları çıkarın

# Korelasyon matrisini oluşturma
correlation_matrix = df_cleaned.corr()

# Korelasyon matrisini görselleştirme
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title('Korelasyon Matrisi')
plt.show()
```

```python
correlation_matrix
```

```python
# Yaşın dağılımını görselleştirelim
plt.figure(figsize=(10, 6))
sns.histplot(df['age'], kde=True, color='skyblue')
plt.title('Yaş Dağılımı')
plt.xlabel('Yaş')
plt.ylabel('Frekans')
plt.show()
```

```python
# Yaş aralıklarına göre gruplandırma
age_bins = pd.cut(df['age'], bins=[0, 10, 20, 30, 40, 50 , 60 , 70], right=False)
age_distribution = age_bins.value_counts().sort_index()

print(age_distribution)
```

```python
# Bölge kategorilerine göre churn riskini görselleştirelim
plt.figure(figsize=(8, 5))
sns.countplot(x='region_category', hue='churn_risk_score', data=df, palette='coolwarm')
plt.title('Bölge Kategorilerine Göre Churn Risk Dağılımı')
plt.xlabel('Bölge Kategorisi')  # Region category axis label updated
plt.ylabel('Churn Sayısı')  # Count axis label remains the same
plt.xticks(ticks=[0, 1, 2], labels=['Town', 'City', 'Village'])  # Updated x-axis labels
plt.show()
```

```python
# Bölge kategorilerine ve churn risk skorlarına göre frekans tablosu
region_churn_summary = df.groupby(['region_category', 'churn_risk_score']).size().unstack()
print(region_churn_summary)
```

```python
# Bölge kategorilerine göre yüzdesel dağılım
region_churn_percentage = region_churn_summary.div(region_churn_summary.sum(axis=1), axis=0) * 100
print(region_churn_percentage)
```

```python
# Üyelik kategorisine göre churn riskini görselleştirelim
plt.figure(figsize=(10, 6))
sns.countplot(x='membership_category', hue='churn_risk_score', data=df, palette='coolwarm')
plt.title('Üyelik Kategorisine Göre Churn Risk Dağılımı')
plt.xlabel('Üyelik Kategorisi')
plt.ylabel('Churn Sayısı')
plt.xticks(ticks=[0, 1, 2 , 3 , 4 , 5], labels=['No Membership', 'Gold', 'Basic' , 'Silver' , 'Premium' , 'Platinum'])
plt.show()
```

```python
# Üyelik kategorilerine ve churn risk skorlarına göre frekans tablosu
membership_churn_summary = df.groupby(['membership_category', 'churn_risk_score']).size().unstack()
print(membership_churn_summary)
```

```python
# Üyelik kategorilerine göre yüzdesel dağılım
membership_churn_percentage = membership_churn_summary.div(membership_churn_summary.sum(axis=1), axis=0) * 100
print(membership_churn_percentage)
```

```python
#Üyelik Kategorisine Göre Ortalama Harcanan Zaman
plt.figure(figsize=(12, 6))
sns.barplot(x='membership_category', y='avg_time_spent', data=df, palette='muted', hue='membership_category')
plt.title('Üyelik Kategorisine Göre Ortalama Harcanan Zaman')
plt.xlabel('Üyelik Kategorisi')
plt.ylabel('Ortalama Zaman (Dakika)')
plt.xticks(ticks=[0, 1, 2, 3, 4, 5], labels=['No Membership', 'Gold', 'Basic' , 'Silver' , 'Premium' , 'Platinum'])
plt.legend(title='Üyelik Kategorisi')
plt.show()
```

```python
# Üyelik kategorilerine göre ortalama harcanan zaman
membership_avg_time = df.groupby('membership_category')['avg_time_spent'].mean()
print(membership_avg_time)
```

```python
# Her üyelik kategorisinin toplam harcanan zamana katkısı
membership_total_time = df.groupby('membership_category')['avg_time_spent'].sum()
total_time = membership_total_time.sum()

membership_time_percentage = (membership_total_time / total_time) * 100
print(membership_time_percentage)
```

```python
# Cüzdan puanlarına göre churn riski
plt.figure(figsize=(10, 6))
sns.barplot(
    x=pd.cut(df['points_in_wallet'], bins=5, labels=["Çok Düşük", "Düşük", "Orta", "Yüksek", "Çok Yüksek"]),
    y='churn_risk_score',
    data=df,
    hue=pd.cut(df['points_in_wallet'], bins=5, labels=["Çok Düşük", "Düşük", "Orta", "Yüksek", "Çok Yüksek"]),
    palette="cool",
    errorbar=None
)
plt.title('Cüzdan Puanlarına Göre Ortalama Churn Riski', fontsize=14)
plt.xlabel('Cüzdan Puanı Kategorileri')
plt.ylabel('Ortalama Churn Riski')
plt.show()
```

```python
# Cüzdan puanlarını kategorilere ayıralım
df['points_category'] = pd.cut(df['points_in_wallet'], bins=5, labels=["Çok Düşük", "Düşük", "Orta", "Yüksek", "Çok Yüksek"])

# Her bir kategori için ortalama churn riskini hesaplayalım
points_churn_risk = df.groupby('points_category', observed=True)['churn_risk_score'].mean()
points_churn_risk
```

```python
# Ortalama giriş sıklığı ve özel indirim kullanımı
plt.figure(figsize=(8, 5))
sns.barplot(
    x='used_special_discount',
    y='avg_frequency_login_days',
    hue='used_special_discount',  # x'i hue olarak belirtiyoruz
    data=df,
    errorbar=None,
    palette="cividis",
    legend=False  # Gereksiz uyarıları önlemek için
)
plt.title('Özel İndirim Kullanımına Göre Ortalama Giriş Sıklığı', fontsize=14)
plt.xlabel('Özel İndirim Kullanımı (0: Hayır, 1: Evet)')
plt.ylabel('Ortalama Giriş Sıklığı (Gün)')
plt.show()
```
